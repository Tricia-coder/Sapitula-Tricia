```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# Uploading the CSV file (ds_salaries.csv)
uploaded = files.upload()
```

```python
# Reading the uploaded CSV file into a DataFrame
df = pd.read_csv("ds_salaries (2).csv")

# Displaying the first few rows to understand the data
df.head()
```

```python
# Printing all unique job titles from the dataset
unique_titles = df['job_title'].unique()
print("Unique Job Titles:")
print(unique_titles)
```

```python
# Grouping by job title and calculating average, standard deviation, and total salary
salary_stats = df.groupby('job_title')['salary_in_usd'].agg(['mean', 'std', 'sum']).reset_index()

# Formatting with dollar signs and rounding
salary_stats['mean'] = salary_stats['mean'].apply(lambda x: f"${x:,.2f}")
salary_stats['std'] = salary_stats['std'].apply(lambda x: f"${x:,.2f}")
salary_stats['sum'] = salary_stats['sum'].apply(lambda x: f"${x:,.2f}")

# Renaming columns for clarity
salary_stats.columns = ['Job Title', 'Average Salary (USD)', 'Standard Deviation (USD)', 'Total Cost (USD)']

# Displaying the final statistics
print("Salary Statistics by Job Title:")
print(salary_stats)
```

```python
# Calculating the average salary for Data Scientists
ds_avg = df[df['job_title'] == 'Data Scientist']['salary_in_usd'].mean()
print("Average Salary (USD) of Data Scientists:")
print(f"${ds_avg:,.2f}")

```

```python
# Calculating the standard deviation for Data Scientists
ds_std = df[df['job_title'] == 'Data Scientist']['salary_in_usd'].std()
print("Standard Deviation of Salary (USD) of Data Scientists:")
print(f"${ds_std:,.2f}")

```
